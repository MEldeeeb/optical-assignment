```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
```

```python
v =sy.symbols('v')                                  # creating a sembole by sympy 
bv = v*(1-((1+2**0.5)**2/(1+(4+v**4)**0.25)**2))    # defining the function we are getting the drivitive for 
dbv = sy.diff(bv,v)                                 # first drivitive of bv with respect to v
ddbv = sy.diff(dbv,v)                               # seconde drivitive of bv with respect to v
print (dbv)
print(ddbv)
#numinator = (1+2**0.5)**2
#denuminator = (1+(4+v**4)**0.25)**2
```

```python
x = np.arange(0,3,0.01)      # creating a list of nums that has start value end value and the step
X = list(x)                  # converting the array to list we can ignor this step

b_v=[]                       #
d_b_v=[]                     # creating an empty list 
d_b_v_2=[]                   #

for v in x:                  #looping in the X (list) 
  
  bv = (1-((1+2**0.5)**2/(1+(4+v**4)**0.25)**2))
  dbv = 11.6568542494924*v**4*(v**4 + 4)**(-0.75)/((v**4 + 4)**0.25 + 1)**3 + 1 - 5.82842712474619/((v**4 + 4)**0.25 + 1)**2
  ddbv = v*(-34.9705627484771*v**7*(v**4 + 4)**(-1.75)/((v**4 + 4)**0.25 + 1)**3 - 34.9705627484771*v**7*(v**4 + 4)**(-1.5)/((v**4 + 4)**0.25 + 1)**4 + 58.2842712474619*v**3*(v**4 + 4)**(-0.75)/((v**4 + 4)**0.25 + 1)**3)

  b_v.append(bv)             # appending the calculated value of bv for each value in X to b_v[]
  d_b_v.append(dbv)          # appending the calculated value of dbv for each value in X to d_b_v[]
  d_b_v_2.append(ddbv)       # appending the calculated value of ddbv for each value in X to d_b_v_2[]

plt.figure(figsize=(9,5))    # creating the figure that we will plot the diagram on it 
plt.plot(X,b_v)              # poting the first diagram
plt.plot(X,d_b_v)            # poting the seconde diagram
plt.plot(X,d_b_v_2)          # poting the thired diagram
plt.title('Mohamed Mahamoud Mahamoud Eldeeb')
plt.xlabel("v_parameter")
```
